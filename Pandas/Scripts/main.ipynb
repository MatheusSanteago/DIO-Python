```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../Data/Netflix TV Shows and Movies.csv')
```

```python
dataset_cut = dataset[['title', 'type', 'release_year', 'age_certification', 'runtime', 'imdb_score', 'imdb_votes']]
```

```python
dataset_cut.loc[:, 'type'] = dataset_cut.loc[:, 'type'].str.capitalize()
```

```python
df = dataset_cut.dropna()
```

```python
df['imdb_score'].plot(kind='hist', bins=10, edgecolor='black')
plt.title('Histograma de Pontuação IMDb')
plt.xlabel('Pontuação IMDb')
plt.ylabel('Frequencia')
plt.show()
```

```python
df.plot(kind='scatter', x='release_year', y='runtime', color='green')
plt.title('Diagrama de Dispersão: Ano de Lançamento vs. Duração')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Duração (minutos)')
plt.show()
```

```python
import seaborn as sns
# Criar o DataFrame com os dados fornecidos
# (Certifique-se de que df já está definido)

# Configurar o estilo de fundo do Seaborn
sns.set(style="whitegrid")

# Criar um gráfico de barras horizontal com cores diferenciadas para os tipos de conteúdo
plt.figure(figsize=(10, 6))
sns.barplot(x='imdb_score', y='title', hue='type', data=df, palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('Comparação de Pontuações IMDb para Filmes e Programas de TV')
plt.xlabel('Pontuação IMDb')
plt.ylabel('Título')

# Adicionar uma legenda
plt.legend(title='Tipo', loc='lower right')

# Mostrar o gráfico
plt.show()
```
